```python
import boto3
import pandas as pd
import gzip
import io
# Specify the S3 bucket and folder path
bucket_name = 'abiomed-hr-pci'
# folder_path = 'outbound/SOW2A/20230117/DIAGNOSES/'
folder_path = 'outbound/SOW2A/20230117/MEDADMINS/'

# Initialize an S3 client
s3 = boto3.client('s3')

# Get a list of all files in the folder
response = s3.list_objects_v2(Bucket=bucket_name, Prefix=folder_path)

if 'Contents' in response:
    file_list = [obj['Key'] for obj in response['Contents'] if obj['Key'].endswith('.csv.gz')]
    
    # Initialize an empty list to hold the dataframes
    df_list = []

    # Loop through each file and read it into a dataframe
    for file_path in file_list:
        # print(f"Reading file: {file_path}")
        # Download the compressed file from S3 and read it into memory
        response = s3.get_object(Bucket=bucket_name, Key=file_path)
        content = response['Body'].read()
        data = gzip.decompress(content)

        # Convert the CSV data to a Pandas dataframe and add it to the list
        df_list.append(pd.read_csv(io.StringIO(data.decode('utf-8')), sep='|'))

    # Concatenate the dataframes into a single dataframe
    df = pd.concat(df_list, ignore_index=True)
else:
    print("No files found in the specified S3 folder.")
```
